```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
scores = pd.read_csv('scores.csv')
DF = 1-np.array(scores['DF'])
LLF = 1-np.array(scores['LLF'])
OrigLLF = 1-np.array(scores['OrigLLF'])
scores.head()
```

```python
scores_ln = pd.read_csv('scores_ln.csv')
DF_ln = 1-np.array(scores_ln['DF'])
LLF_ln = 1-np.array(scores_ln['LLF'])
OrigLLF_ln = 1-np.array(scores_ln['OrigLLF'])

scores_ln.head()
```

```python
sns.set()

c=sns.color_palette('Set1')

fs=25
ls=22.5
fig, ax = plt.subplots(1,1, figsize=(8,6))
ax.plot(DF[0] / OrigLLF, label='Lifelong Forest - Task 1', lw=3, c=c[0])
ax.plot(DF_ln[0] / OrigLLF_ln[:79], label='Lifelong Network - Task 1', lw=3, c=c[0], ls='--')
# ax.axhline(np.mean(DF / LLF), label='Average FTE', lw=3, c=c[1])
# ax.axhline(np.mean())
ax.legend(loc='upper left', fontsize=18)
ax.set_ylabel('Transfer Efficiency', fontsize=fs)
ax.set_xlabel('Number of tasks seen', fontsize=fs)
ax.set_title('Random Subsets of 10 Classes', fontsize=fs)
# ax.set_ylim(1-0.01, 1.2+0.01)
ax.set_yticks([1, 1.1, 1.2, 1.3])
ax.tick_params(labelsize=ls)
plt.tight_layout()
plt.savefig('random_subsets_of_10_classes_with_LN.pdf')
```
